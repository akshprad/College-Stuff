```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds=pd.read_csv('Mall_Customers.csv')
ds.head()
```

```python
X=ds.iloc[:, [3,4]].values
X
```

```python
#Scaling Data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
#Applying K-Means to our dataset with 5 centroids
from sklearn.cluster import KMeans
km_final=KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_means= km_final.fit_predict(X)
```

```python
#Plotting the cluster plots
plt.scatter(X[y_means==0, 0], X[y_means== 0, 1], c='red', s=50, label='c1')
plt.scatter(X[y_means==1, 0], X[y_means== 1, 1], c='blue', s=50, label='c2')
plt.scatter(X[y_means==2, 0], X[y_means== 2, 1], c='green', s=50, label='c3')
plt.scatter(X[y_means==3, 0], X[y_means== 3, 1], c='orange', s=50, label='c4')
plt.scatter(X[y_means==4, 0], X[y_means== 4, 1], c='yellow', s=50, label='c5')

#the pink points are the centroids
plt.scatter(km_final.cluster_centers_[:, 0], km_final.cluster_centers_[:, 1], c='magenta', s=150)

plt.legend()
plt.show()
```
